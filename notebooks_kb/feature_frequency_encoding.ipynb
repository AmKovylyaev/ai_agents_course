# Frequency encoding

```python
def add_frequency_encoded_features(train_df, test_df, columns):
    train_df = train_df.copy()
    test_df = test_df.copy()
    for col in columns:
        freq = train_df[col].value_counts(dropna=False).to_dict()
        train_df[f"{col}_freq"] = train_df[col].map(freq).fillna(0)
        test_df[f"{col}_freq"] = test_df[col].map(freq).fillna(0)
    return train_df, test_df
```
