# EDA and target detection

```python
import pandas as pd

train = pd.read_csv("data/train.csv")
sample = pd.read_csv("data/sample_submition.csv")

possible_targets = [c for c in train.columns if c not in sample.columns]
target_column = possible_targets[-1] if possible_targets else train.columns[-1]

numeric_columns = train.drop(columns=[target_column], errors="ignore").select_dtypes(include=["number"]).columns.tolist()
categorical_columns = train.drop(columns=[target_column], errors="ignore").select_dtypes(exclude=["number"]).columns.tolist()

target = train[target_column]
if target.dtype.kind in "biu" and target.nunique(dropna=True) <= 20:
    task_type = "classification"
    n_classes = int(target.nunique(dropna=True))
else:
    task_type = "regression"
    n_classes = 1

eda_summary = {
    "target_column": target_column,
    "numeric_columns": numeric_columns,
    "categorical_columns": categorical_columns,
    "task_type": task_type,
    "n_classes": n_classes,
}
print(eda_summary)
```
