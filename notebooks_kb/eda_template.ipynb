# EDA template for tabular competition

```
import pandas as pd
from pathlib import Path

DATA_DIR = Path("data")
train = pd.read_csv(DATA_DIR / "train.csv")
test = pd.read_csv(DATA_DIR / "test.csv")

print("train shape:", train.shape)
print("test shape:", test.shape)
print(train.head())
```

## Missing values and dtypes

```
missing = train.isna().mean().sort_values(ascending=False)
print(missing.head(20))

print(train.dtypes.value_counts())
```

## Cardinality of categorical features

```
cat_cols = train.select_dtypes(include=["object", "string", "category"]).columns.tolist()
cardinality = {col: train[col].nunique(dropna=False) for col in cat_cols}
print(dict(sorted(cardinality.items(), key=lambda x: x[1], reverse=True)[:20]))
```
